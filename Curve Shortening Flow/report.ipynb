# Отчёт

```python
import numpy as np
import matplotlib.pyplot as plt
import curve_shortening_flow as csf

%matplotlib inline
```

```python
curves = [[(0., 0.), (1., 0.), (1., 1), (0., 1.), (0.5, 0.5), (0., 0.5)],
          [(1., 0.5), (0.75, 0.5 + np.sqrt(0.75) * 0.5), (0.5, 1.), (0., 0.5), (0.5, 0.)]]

flows = [{v: csf.CurveShorteningFlow(curve, v) for v in ['A', 'B', 'C', 'D']} for curve in curves]
```

```python
def test(flow):
    mass_center = flow.get_mass_center()
    total_curvature = flow.get_total_curvature()
    is_round = flow.is_round()

    flow.transform()

    new_mass_center = flow.get_mass_center()
    new_total_curvature = flow.get_total_curvature()
    new_is_round = flow.is_round()

    print("Center of masses saved?: ", np.isclose(mass_center, new_mass_center))
    print("Total curvature saved?: ", np.isclose(total_curvature, new_total_curvature))
    print("Roundness saved?: ", is_round and new_is_round)
```

# Кривая 1, Вариант A

```python
test(flows[0]['A'])
```
