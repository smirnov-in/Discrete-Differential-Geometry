# Deep Composing

Original paper: https://graphics.pixar.com/library/DeepCompositing/paper.pdf

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
from image import Color, DeepImage, Image
```

```python
smoke_1 = Image.read('smoke_1.png')
smoke_1.resize((600, 600))
smoke_1 *= Color([1., .455, 0.], .99)
```

```python
smoke_1.show()
```

```python
smoke_2 = Image.read('smoke_2.png')
smoke_2.resize((600, 600))
smoke_2 *= Color([.455, 0., 1.], .8)
```

```python
smoke_2.show()
```

```python
smoke_3 = Image.read('smoke_3.png')
smoke_3.resize((600, 600))
smoke_3 *= Color([0., 1., 0.])
```

```python
smoke_3.show()
```

```python
depths_1 = smoke_1.data[:, :, -1]
depths_2 = smoke_2.data[:, :, -1]
depths_3 = smoke_3.data[:, :, -1]

shift = np.tile(np.linspace(1, -1, 600), (600, 1))

d_smoke_1 = DeepImage(smoke_1, -depths_1/2, depths_1/2)
d_smoke_2 = DeepImage(smoke_2, -depths_2/2 + shift, depths_2/2 + shift)
d_smoke_3 = DeepImage(smoke_3, -depths_3/2, depths_3/2)
```

```python
d_smoke_1.render().show()
d_smoke_1.render(Color.black()).show()
```

```python
d_smoke_2.render().show()
d_smoke_2.render(Color.black()).show()
```

```python
d_smoke_3.render().show()
d_smoke_3.render(Color.black()).show()
```

```python
(d_smoke_1 * d_smoke_2).render().show()
```

```python
(d_smoke_1 * d_smoke_3).render().show()
```

```python
(d_smoke_2 * d_smoke_3).render().show()
```

```python
(d_smoke_1 * d_smoke_2 * d_smoke_3).render().show()
```
